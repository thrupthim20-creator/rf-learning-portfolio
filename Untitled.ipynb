# Project 2 — FFT Spectrum Analyser

## What this project does
Loads a real audio WAV file and analyses its frequency content using FFT.
Visualises the signal in both time domain and frequency domain.

## Concepts covered
- Reading real-world audio data into Python
- FFT on a real signal (not synthetically generated)
- Identifying frequency components of a real sound

## Tools used
- NumPy, Matplotlib, SciPy

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
# scipy.io.wavfile: module specifically for reading and writing WAV audio files
```

```python
fs, data = wavfile.read("audio.wav")
if data.ndim == 2:
    data = data[:,0]

data = data.astype(np.float32)

print(f"Sampling rate: {fs} Hz")
print(f"Number of samples: {len(data)}")
print(f"Duration: {len(data)/fs:.2f} seconds")
```

```python
t = np.arange(len(data)) * 1/fs

plt.plot(t, data, linewidth=0.5)
# linewidth=0.5: thin line so individual samples are visible on a dense signal

plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Audio Signal — Time Domain")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
fft = np.fft.fft(data)
freq = np.fft.fftfreq(len(data), 1/fs)

magnitude = np.abs(fft)

plt.figure(figsize=(10, 4))
plt.plot(freq, magnitude, linewidth=0.5)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("Audio Signal — Frequency Spectrum")
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# most audio content sits below 8000 Hz — zoom in to see it clearly
plt.figure(figsize=(10, 4))
plt.plot(freq, magnitude, linewidth=0.5)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("Frequency Spectrum — zoomed to 0–8000 Hz")
plt.xlim(0, 8000)
plt.xticks(ticks=[100,300,500,1000,2000,3000,4000,5000,6000,7000,8000], rotation=45)
# plt.xlim(): sets the x-axis limits so we zoom into 0 to 8000 Hz only
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Key observations

- The frequency range of the audio signal is below 10000 Hz
- The strongest spike is at 300 Hz, that's likely the fundamental frequency of the sound. If it's a voice, that's the pitch of the speaker
- Compared to sine wave, the frequency is not concetrated at one point but its distributed from 0 to 10000 Hz 

## What I learned
- Real signals contain many frequency components, not just one spike
- FFT works the same way on real audio as on synthetic signals
- Zooming the frequency axis helps identify specific components
